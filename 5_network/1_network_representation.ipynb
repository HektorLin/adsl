```python
import networkx as nx
```

```python
## Create a directed graph. What would be the command for an undirected graph?
G = nx.DiGraph()
```

```python
## Add nodes.
G.add_nodes_from([1,2,3,4,5])
```

```python
## Print all nodes.
G.nodes()
```

Try to represent the network in the slides as adjacency list first, then move to its visualization. 
Then repeat the same step with the edge list. 

#### Adjacency list.
Write the adjacency as a list of lists.

```python
# Your code here.
A = [
    [0,1,1,1,1],
    [1,0,1,1,1],
    [1,1,0,1,1],
    [1,1,1,0,1],
    [1,1,1,1,0]
    ]
```

The code below uses `add_adge` to create a add a edge to the graph.

```python
## Get the number of nodes.
N = G.number_of_nodes()

## Iterate through the adjancency matrix.
for i in range(N):
    for j in range(N):
        link = A[i][j]
        if link != 0:
            # print('Adding', i, j)
            G.add_edge(i+1, j+1)
```

#### Edge list.
Write the edge list as a list of lists.

```python
# Your code here.
E = [[1,2],[3,4]]
```

The code below uses `add_adge` to create a add a edge to the graph.

```python
## Get the number of nodes.
N = G.number_of_nodes()

## Iterate through the edge list.
for i, edges in enumerate(E):
    for j in range(len(edges)):
        # print('Adding', i+1, edges[j])
        G.add_edge(i+1, edges[j])
```

```python
## Make a larger figure.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

# Set a layout.
pos = nx.circular_layout(G)

## With larger node size (default=300) and some style.
nx.draw(G, pos, with_labels=True, node_size=2000, 
                font_size=20, linewidths=2, node_color="lightblue")
```
